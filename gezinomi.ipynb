```python
# SaleCheckInDayDiff değişkenini kategorik bir değişkene çeviriniz.
# ‘0_7’, ‘7_30', ‘30_90', ‘90_max’ aralıklarını kullanabilirsiniz.
# Bu aralıklar için "Last Minuters", "Potential Planners", "Planners", "Early Bookers“ isimlerini kullanabilirsiniz.

bins = [0, 7, 30, 90, float("inf")]
labels = ["Last Minuters", "Potential Planners", "Planners", "Early Bookers"]

df["Category"] = pd.cut(df['SaleCheckInDayDiff'], bins=bins, labels=labels, right=False, include_lowest=True, duplicates="raise")

df.head(10)

#pd.cut() fonksiyonu belirli bir sayısal değişkeni kategorik bir değişkene dönüştürmek için kullanılır.
"""Parameters:
* x: Dönüştürülecek sayısal değişken.
* bins: Aralıkların sınırlarını belirleyen bir liste veya sayısal bir değer. Aralıkların sınırları, bins parametresine göre belirlenir.
Örneğin, [0, 10, 20, 30] gibi bir liste, aralıkları 0-10, 10-20 ve 20-30 olarak oluşturur.
* labels: Aralıklara atanacak etiketlerin bir listesi. Bu parametre isteğe bağlıdır. Belirtilmezse, aralıkların indeks değerleri kullanılır.
* right: Aralık sınırlarının sağda kapalı (True) veya sağda açık (False) olacağını belirler. Varsayılan değeri True'dur.
* include_lowest: Aralığın alt sınırının dahil edilip edilmeyeceğini belirler. Varsayılan değeri False'dir.
* duplicates: Eşit değerlere sahip birden fazla sınıfın bulunması durumunda hata mesajı ("raise"),
her bir sınıfı farklı bir şekilde işaretleme ("drop") veya eşitlik durumunu koruma ("raise") seçeneklerinden birini belirler.
Varsayılan değeri "raise"'dir.
"""
```

```python
# Şehir, konsept, kategori kırılımında ortalama ödenen ücret ve yapılan işlem sayısı cinsinden inceleyiniz.
df.groupby(by=["SaleCityName", 'ConceptName', "Category" ]).agg({"Price": ["mean", "count"]})
```

```python
# Şehir, konsept, sezon kırılımında ortalama ödenen ücret ve yapılan işlem sayısı cinsinden inceleyiniz.
df.groupby(by=["SaleCityName", 'ConceptName', "Seasons" ]).agg({"Price": ["mean", "count"]})
```

```python
# Şehir, konsept, CInDay kırılımında ortalama ödenen ücret ve yapılan işlem sayısı cinsinden inceleyiniz.
df.groupby(by=["SaleCityName", 'ConceptName', "CInDay" ]).agg({"Price": ["mean", "count"]})
```

```python
# City, Concept, Season kırılımını price'a göre sıralayınız.
data_agg = df.groupby(["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": "mean"}).sort_values("Price",ascending=False)
```

```python
# City, Concept, Season kırılımını price'a göre sıralanmış veriyi agg_df olarak kaydediniz.
agg_df = pd.DataFrame(data = data_agg)
agg_df

# Bunun yerine direkt agg_df = df.groupby(["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": "mean"}).sort_values("Price",ascending=False)
# işlemi de yapılabilir, daha kısa olacaktır.
```

```python
# agg_df için indexte yer alan isimleri değişken ismine çeviriniz.
agg_df.reset_index(inplace=True)
```

```python
# Yeni seviye tabanlı satışları tanımlayınız ve sales_level_based sütununa kaydediniz. Örnek bir değer: GIRNE_HERŞEY DAHIL_HIGH
agg_df["sales_level_based"] = agg_df[["SaleCityName", "ConceptName", "Seasons"]].agg(lambda x: '_'.join(x),axis=1)
```

```python
agg_df.head()
```

```python
# Yeni personaları PRICE’a göre 4 segmente ayırınız.
# Segmentleri SEGMENT isimlendirmesi ile değişken olarak agg_df’e ekleyiniz.
agg_df["Segment"] = pd.qcut(agg_df["Price"], 4, labels=["D","C","B","A"], duplicates="raise")
# Segmentleri betimleyiniz (Segmentlere göre group by yapıp price mean, max, sum’larını alınız)
agg_df.groupby(agg_df["Segment"]).agg({"Price" : ["mean", "max", "sum"]})
```

```python
# Antalya’da herşey dahil ve yüksek sezonda tatil yapmak isteyen bir kişinin ortalama ne kadar gelir kazandırması beklenir?
new_customer = "Antalya_Herşey Dahil_High"
agg_df[agg_df['sales_level_based'] == new_customer]
```

```python
# Girne’de yarım pansiyon bir otele düşük sezonda giden bir tatilci hangi segmentte yer alacaktır?
agg_df["Segment"][agg_df["sales_level_based"] == "Girne_Yarım Pansiyon_Low"]
```
